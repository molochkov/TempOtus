```python
import numpy as np
import numpy.linalg as la
import matplotlib.pyplot as plt

n = 200
matrix = np.random.randint(0, 10, (n, n))
U, s, VT = la.svd(matrix)
ranks = range(2, n + 1)

errors = []

for r in ranks:
    sigma = np.diag(s)
    mat_approx = U[:, :r] @ sigma[:r, :r] @ VT[:r, :]
    error = np.linalg.norm(matrix - mat_approx)
    errors.append(error)


plt.plot(ranks, errors)
plt.xlabel('Ранг r')
plt.ylabel('Ошибка аппроксимации')
plt.title('Зависимость ошибки аппроксимации от ранга')
plt.show()
```

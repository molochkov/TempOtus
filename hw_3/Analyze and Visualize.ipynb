### Создайте массив 4x3x2 со случайными значениями. Найти его минимум и максимум.

```python
a = np.random.randint(0, 100, size=(4, 3, 2))
max = a.max()
min = a.min()

print(f'max = {max}')
print(f'min = {max}')
```

### Создайте два двумерных массива размерами 6x4 и 4x3 и произведите их матричное умножение. 

```python
a = np.random.randint(0, 100, size=(6, 4))
b = np.random.randint(0, 100, size=(4, 3))

result = np.dot(a,b)
result
```

### Создайте случайный двумерный массив 7x7, найти у него среднее и стандартное оклонение. Нормализуйте этот массив.

```python
a = np.random.randint(0, 100, size=(7, 7))
av = np.average(a)
std = np.std(a)
print(f'Среднее отклонение : {av}')
print(f'Стандартное отклонение : {std}')

norm=np.linalg.norm(a)
normalize = a/np.linalg.norm(a)
print(normalize)
```

### Узнайте какой средний 'total_bill' в зависимости от 'day'

```python
day_avg = tips_data[['day','total_bill']].groupby('day').mean()
day_avg
```

# III. Visualization

### Постройте гистограмму распределения признака 'total_bill'

```python
sns.displot(tips_data['total_bill'])
```

### Постройте scatterplot, представляющий взаимосвязь между признаками 'total_bill' и 'tip'

```python
import matplotlib.pyplot as plt
plt.scatter(tips_data.total_bill, tips_data.tip, alpha=0.8,
            cmap='viridis')
plt.title("Взаимосвязь между признаками 'total_bill' и 'tip'")
plt.xlabel("total_bill")
plt.ylabel("tip")
plt.colorbar()
plt.show()
```

### Постройте pairplot

```python
sns.pairplot(tips_data);
```

### Постройте график взаимосвязи между признаками 'total_bill' и 'day'

```python
m= {'Mon': 1, 'Tue': 2,'Wed': 3, 'Thur': 4,'Fri': 5, 'Sat': 6,'Sun': 7}
df = tips_data[['total_bill','day']].copy()
df['day']= df['day'].map(m)
lb = ['Mon','Tue','Wed','Thur','Fri','Sat','Sun']
digits = [1,2,3,4,5,6,7]
plt.scatter(df.day,df.total_bill)
plt.xticks(digits,lb)
```

### Постройте два графика scatterplot, представляющих взаимосвязь между признаками 'total_bill' и 'tip' один для Male, другой для Female и раскрасьте точки в зависимоти от признака 'smoker'

```python
dfMale = tips_data[tips_data.sex=='Male']
dfFemale = tips_data[tips_data.sex=='Female']

df = tips_data.copy()
df['smoker']= df['smoker'].map({'Yes': 0, 'No': 1})

dfMale = df[df.sex=='Male']
dfFemale = df[df.sex=='Female']


plt.scatter(dfMale.total_bill, dfMale.tip, alpha=0.8, c=dfMale.smoker, cmap='viridis')
plt.title("Взаимосвязь между признаками 'total_bill' и 'tip' для Male")
plt.xlabel("total_bill")
plt.ylabel("tip")
plt.show()

plt.scatter(dfFemale.total_bill, dfFemale.tip, alpha=0.8,c=dfFemale.smoker, cmap='viridis')
plt.title("Взаимосвязь между признаками 'total_bill' и 'tip' для Female")
plt.xlabel("total_bill")
plt.ylabel("tip")
plt.show()
```
